1. ¿Por qué una matriz equivale a una transformación lineal entre espacios vectoriales?
> Cualquier transformación lineal $\ T: V → W$ puede representarse mediante una matriz: $\ T(x) = A x$, en la cual la matriz A dependerá de las bases elegidas para V y W.   
La matriz de una transformación lineal queda determinada cuando se conocen las bases de sus espacios vectoriales, en éste caso: V, y W)  
Suponiendo que el espacio vectorial V tiene una base {v1, ..., vn} y el espacio vectorial W tiene una base {w1, ..., wm}, entonces cualquier transformación lineal de V en W se representa por una matriz $\ A_\mathtt{mxn} $.  
Si T (vi ) = ai1 w1 + .... + aim wm, entonces la columna i de A es (ai1 .... aim )T
Toda transformación lineal $\ T: R^n → R^m $ es una transformación matricial.  
Trabajar con una transformación lineal es equivalente a trabajar con la matriz de la transformación.  

+ ¿Cuál es el efecto de transformación lineal de una matriz diagonal y el de una matriz ortogonal?  
> La transformación lineal sobre una matriz diagonal tiene le efecto de dilatar o contraer los ejes, mientras que la transformación sobre una matriz ortogonal produce rotación y/o reflexión.

+ ¿Qué es la descomposición en valores singulares de una matriz?
> La SVD de una matriz A es una factorización de la forma: $\ A = UΣV^T$, donde:  
>* U → matriz con columnas ortogonales de m × n  
>* V → es una matriz ortogonal de n × n   
>* Σ → una matriz “diagonal”de m × n. Con $\ σ_{11} >= σ_{22} >= .... >= σ_{mn} >= 0 $ 
>* donde las columnas de U y V son vectores singulares y los $\ σ_{ii} $ de Σ son valores singulares.

+ ¿Qué es diagonalizar una matriz y que representan los eigenvectores?
> Diagonalizar una matriz A es encontrar una matriz diagonal semejante a A.  
Una matriz A de n x n es diagonalizable si existe una matriz diagonal D tal que A es semejante a D. A es diagonalizable si y sólo si tiene **n** eigenvectores linealmente independientes. En tal caso, la matriz diagonal D semejante está dada por:  
\begin{equation}
D = 
\left(\begin{array}{cc} λ_1 & ... & 0\\ ... & ... & ... \\ 0 & ... & λ_n \end{array}\right)
\end{equation}
> donde los eigenvectores representan las culumnas.  
> Matrices semejantes:
>* Se dice que dos matrices A y B de n x n son semejantes si existe una matriz invertible C de n x n tal que: $\ B = CAC^\mathtt{-1}$  
>* A y B son semejantes si y sólo si existe una matriz no singular C tal que CB = AC  
>* Las matrices semejantes tienen los mismos valores característicos.  

+ ¿Intuitivamente qué son los eigenvectores?
> Los eigenvectores son los ejes principales los cuales preservan su dirección.  

+ ¿Cómo interpretas la descomposición en valores singulares como una composición de tres tipos de transformaciones lineales simples?
> La SVD puede interpretarse como 3 transformaciones lineales simples $\ A = UΣV^T$
>* U → rotación
>* Σ → escalamiento
>* V → rotación

+ ¿Qué relación hay entre la descomposición en valores singulares y la diagonalización?
> La SVD es una generelización de la diagonalización.  
> En ambos casos la matriz puede interpretarse como 3 transformaciones lineales simples:  
> * SDV: $\ A = UΣV^T$  
> * Diagonalización: $\ A = PAP^\mathtt{-1}$   

+ ¿Cómo se usa la descomposición en valores singulares para dar un aproximación de rango menor a una matriz?
> Se debe a que en la descomposición SVD la información mas relevante de una matriz se encuentran en los primeros terminos de la descomposición y va asociada a los valores singulares mas grandes lo que permite obtener matices de rango inferior pero muy parecidas a la original.  

+ Describe el método de minimización por descenso gradiente
> Recordando que el gradiente es un vector en un punto x que proporciona la dirección (local) de máxima variación de la función. El vector gradiente es un vector ortogonal al contorno de la función en el punto.    
Por lo tanto en la búsqueda de un mínimo la dirección de movimiento será contragradiente: $\ -∇f(X)$  
> Ee método de minimzación por descenso de gradiente es un proceso iterativo de optimizacion sin restricciones.
Para ello se deben llevar a cabo los siguientes pasos:  
> * 1- Elegir un punto inicial $\ x_k$, k = 0  
> * 2- Calcular $\ -∇f(x_k)$   
> * 3- Determinar el punto siguiente: $\ x_\mathtt{k+1} = x_k -α∇f(x_k)$    
> donde α es el tamaño del paso el cual puede ser un valor fijo o determinarse heurísticamente.  
> * 4- Comparar $\ f(x_\mathtt{k+1})$ con $\ f(x_k)$   
> Si el cambio es menor que una tolerancia preespecificada, termina el proceso, en caso contrario volver al paso 3 y continuar con las iteraciones.

#### 2.2. Ahora veremos la aplicación a pseudoinversa y sistemas de ecuaciones:
* Programar una función que dada cualquier matriz devuelva la pseudainversa usando la descomposición SVD.  
 Hacer otra función que resuelva cualquier sistema de ecuaciones de la forma Ax=b usando esta pseudoinversa.
* Jugar con el sistema Ax=b donde A=[[1,1],[0,0]] y b puede tomar distintos valores. 
    * (a) Observar que pasa si b esta en la imagen de A (contestar cuál es la imagen) y si no está (ej. b = [1,1]).
    * (b) Contestar, ¿la solución resultante es única? Si hay más de una solución, investigar que carateriza a la solución devuelta. 
    * (c) Repetir cambiando A=[[1,1],[0,1e-32]], ¿En este caso la solucíon es única? ¿Cambia el valor devuelto de x en cada posible valor de b del punto anterior?
    > Si **b** está en la imagen, la función retorna la solución única del sistema, en cmabio cuando no lo está retorna la solución mas aproximada.

```python
def svd_pseudoinv(pA):
    # A = U * Sigma * V^t
    U, sigma, V = np.linalg.svd(pA)
        
    # sigma_plus = np.diag(np.array([1 / val if val != 0 else 0 for val in sigma]))    
    
    sigma_plus = np.zeros([U.shape[1],V.shape[0]])
    
    nrow, ncol = sigma_plus.shape    
    for i in range(nrow):
        for j in range(ncol):
            if i == j:
                if sigma[i] != 0:
                    sigma_plus[i,j] = 1/sigma[i]

    return np.dot(np.dot(np.transpose(V), np.transpose(sigma_plus)), np.transpose(U))

def svd_solve_ecu(pA,pb):
    # x = A^+ * b
    A_plus = svd_pseudoinv(pA)        
    return np.matmul(A_plus, pb)
```

```python
A1 = np.array([[1,1],[0,0]])
b1 = np.array([1,1])

xa1 = svd_solve_ecu(A1,b1)
xa1
```

```python
A2 = np.array([[1,1],[0,0]])
b2 = np.array([1,0])

xa2 = svd_solve_ecu(A2,b2)
xa2
```

```python
A3 = np.array([[1,1],[0,1e-32]])
b3 = np.array([1,1])

xa3 = svd_solve_ecu(A3,b3)
xa3
```

##### Optimizacion - Regresión lineal

```python
# np.linalg.lstsq resuelve la ecuación "a x = b" 
# calculando un vector x que minimiza la norma_2 ||B - a x||^2

A = np.array([df.study_hours, np.ones(len(df))])

# Calculo coeficientes:
beta,alfa = np.linalg.lstsq(A.T,df.sat_score)[0] # obtaining the parameters

print("Coeficientes:")
print(alfa, beta)
```

##### Predicciones:

```python
def prediccion(df_base, nuevas_hh):
    """Calculo coeficientes para los datos base y luego calcula predicción"""
    A = np.array([df_base.ix[:,0], np.ones(len(df_base))])

    # Calculo coeficientes:    
    b,a = np.linalg.lstsq(A.T,df_base.ix[:,1])[0] # obtaining the parameters
    # Calculo predicción: 
    predicc =  np.array(np.multiply(nuevas_hh,b) + a)
    
    return(a, b, predicc)

def prediccion_coef(pAlfa, pBeta, nuevas_hh):
    """Calculo predicción para los coeficientes indicados"""
    
    predicc =  np.multiply(nuevas_hh,pBeta) + pAlfa
    
    return(np.array(predicc))
```

```python
estimaciones = np.array(df.ix[7.5:,0] + 2.5)
print("Estimaciones: ")
print(str(estimaciones) + "\n")

alfa,beta,predicc = prediccion(df,estimaciones)
print("Coeficientes: " + str(alfa) + " " + str(beta) + "\n")
print("Predicción:")
print(predicc.T)
```

```python
predicc_coef = prediccion_coef(353, 25, estimaciones)
print("Predicción:")
print(predicc_coef)
```
