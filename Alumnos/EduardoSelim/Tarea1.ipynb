```python
class Array:
    "Una clase minima para algebra lineal"    
    def __init__(self, list_of_rows): 
        "Constructor y validador"
        # obtener dimensiones
        self.data = list_of_rows
        nrow = len(list_of_rows)
        #  ___caso vector: redimensionar correctamente
        if not isinstance(list_of_rows[0], list):
            nrow = 1
            self.data = [[x] for x in list_of_rows]
        # ahora las columnas deben estar bien aunque sea un vector
        ncol = len(self.data[0])
        self.shape = (nrow, ncol)
        # validar tamano correcto de filas
        if any([len(r) != ncol for r in self.data]):
            raise Exception("Las filas deben ser del mismo tamano")

#Ejercicio 1: Un metodo para imprimir mejor
        
    def __repr__(self):
        "Para la reconozca con propiedades de matriz"
        s = ""
        for lista in self.data:
            s+=str(lista)
            s+="\n"
        return s
    
    def __str__(self):
        "Para que visualmente para el usuario luzca como matriz"
        s = ""
        for lista in self.data:
            s+=str(lista)
            s+="\n"
        return s

#Ejercicio 2: Acceder a una entrada de la matriz y cambiar una entrada de la matriz

    def __getitem__(self, idx):
        return self.data[idx[0]][idx[1]]
    
    def __setitem__(self, idx, nuevo_valor):
        if (nuevo_valor, (int, float, complex)):
            self.data[idx[0]][idx[1]]=nuevo_valor
        else:
             raise Exception("Las entradas deben ser numericas")

#Ejercicio 4: Transposicion
    
    def transpuesta(self):
        nrow = len(self.data)
        ncol = len(self.data[0])
        trans = Array([[0 for r in range(nrow)] for c in range(ncol)])
        for i in range(nrow):
            for j in range(ncol):
                trans.data[j][i]=self.data[i][j]
        return(trans)
    
    def __add__(self, other):
        "Hora de sumar"
        if isinstance(other, Array):
            if self.shape != other.shape:
                raise Exception("Las dimensiones son distintas!")
            rows, cols = self.shape
            newArray = Array([[0. for c in range(cols)] for r in range(rows)])
            for r in range(rows):
                for c in range(cols):
                    newArray.data[r][c] = self.data[r][c] + other.data[r][c]
            return newArray
        elif isinstance(2, (int, float, complex)): # en caso de que el lado derecho sea solo un numero
            rows, cols = self.shape
            newArray = Array([[0. for c in range(cols)] for r in range(rows)])
            for r in range(rows):
                for c in range(cols):
                    newArray.data[r][c] = self.data[r][c] + other
            return newArray
        else:
            return NotImplemented # es un tipo de error particular usado en estos metodos
        
# Ejercicio 5.1. Para que tenga sentido escalar + Array
    
    __radd__ = __add__
    
# Ejercicio 5.2. Una funcion para resta matricial
    
    def __sub__ (self,other):
        "Hora de restar"
        if isinstance(other, Array):
            if self.shape != other.shape:
                raise Exception("Las dimensiones son distintas!")
            rows, cols = self.shape
            newArray = Array([[0. for c in range(cols)] for r in range(rows)])
            for r in range(rows):
                for c in range(cols):
                    newArray.data[r][c] = self.data[r][c] - other.data[r][c]
            return newArray
        elif isinstance(other, (int, float, complex)): # en caso de que el lado derecho sea solo un numero
            rows, cols = self.shape
            newArray = Array([[0. for c in range(cols)] for r in range(rows)])
            for r in range(rows):
                for c in range(cols):
                    newArray.data[r][c] = self.data[r][c] - other
            return newArray
        else:
            return NotImplemented # es un tipo de error particular usado en estos metodos

# Ejercicio 6.1 Una funcion para multiplicacion matricial 

    def __mul__ (self,other):
        "Hora de mutiplicar"
        if isinstance(other, Array):
            rowsA, colsA = self.shape
            rowsB, colsB= other.shape
            if colsA != rowsB:
                raise Exception("Para poder multiplicar A*B, el número de columnas de A debe coincidir con el número de filas de B")
            
            newArray = Array([[0. for c in range(colsB)] for r in range(rowsA)])
            
            for r in range(rowsA):
                for d in range (colsB):
                    for c in range(colsA):
                        newArray.data[r][d] += self.data[r][c] * other.data[c][d]
            return newArray
        elif isinstance(other, (int, float, complex)): # en caso de que el lado derecho sea solo un numero
            rows, cols = self.shape
            newArray = Array([[0. for c in range(cols)] for r in range(rows)])
            for r in range(rows):
                for c in range(cols):
                    newArray.data[r][c] = self.data[r][c] * other
            return newArray
        else:
            return NotImplemented # es un tipo de error particular usado en estos metodos
        
# Ejercicio 6.2 Para que tenga sentido escalar * Array
    
    __rmul__ = __mul__
    
class Vector(Array): # declara que Vector es un tipo de Array
    def __init__(self, list_of_numbers):
        self.vdata = list_of_numbers
        list_of_rows = [[x] for x in list_of_numbers]
        return Array.__init__(self, list_of_rows)
    def __repr__(self):
        return "Vector(" + str(self.vdata) + ")"
    def __str__(self):
        return str(self.vdata)
    def __add__(self, other):
        new_arr = Array.__add__(self, other)
        return Vector([x[0] for x in new_arr.data]) 
```

```python
# Ejercicio 3
# Funcion matriz de ceros

def MatrizCeros(numren,numcol):
    matrcer=Array([[0. for i in range(numcol)] for j in range(numren)])
    return matrcer
```

```python
# Ejercicio 3
# Funcion matriz identidad

def Eye(n):
    identidad=MatrizCeros(n,n)
    for i in range(n):
        identidad[i,i]=1.0
    return(identidad)
```

```python
# Ejercicio 7
# Solucion de un sistema de ecuaciones triangular inferior

def forward_sub(L,y): # Basado en en el link que nos pasaste
    n,c = L.shape
    x = MatrizCeros(n,1)
    
    if n != c: # Verifica que sea una matriz n x n
        raise Exception("L no es una matriz cuadrada")
    
    for i in range(n): # Verifica que sea una matriz triangular inferior
        for j in range(i+1,n-1):
            if L[i,i +j] != 0:
                raise Exception("L no es una matriz triangular inferior")
    
    for i in range (n):
        x[i,0]= y[i,0]
        
        for j in range(i):
            x[i,0]=x[i,0]-(L[i, j]*x[j,0])
        
        x[i,0] = x[i,0]/L[i, i]
    
    return x
```

```python
# Ejercicio 8
# Solucion de un sistema de ecuaciones triangular superior

def backward_subs(U,y): # Basado en en el link que nos pasaste
    n,c = U.shape
    x = MatrizCeros(n,1)
    
    if n != c: # Verifica que sea una matriz n x n
        raise Exception("U no es una matriz cuadrada")
    
    L=U.transpuesta()
    for i in range(n): # Verifica que sea una matriz triangular inferior
        for j in range(i+1,n-1):
            if L[i,i +j] != 0:
                raise Exception("L no es una matriz triangular superior")    
    
    for i in reversed(range(n)):
        x[i,0]= y[i,0]
        
        for j in range(i+1,n):
            x[i,0]=x[i,0]-(U[i, j]*x[j,0])
        
        x[i,0] = x[i,0]/U[i, i]
    
    return x
```
